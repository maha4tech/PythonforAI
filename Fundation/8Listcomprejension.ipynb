## List Comperhension
List comperhension provides concise way to create lists. it consists of brackets an experssion follwed by a for claus, then zero or more for or if claue

```python
def lst_square(lst):
    for i in lst:
        lst.append(i*i)
```

```python
lst_square([1,2,3,4,5,6,7])
```

```python
lst=([1,2,3,4,5,6,7])
[i*i for i in lst]
```

```python
# for even
[i*i for i in lst if i%2==0]
```

```python
# for odd
[i*i for i in lst if i%2!=0]
```

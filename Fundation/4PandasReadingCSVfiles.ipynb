# Pandas
- Pandas is a popular library in Python used for data analysis and manipulation. It helps you work with structured data like tables (similar to Excel spreadsheets or SQL tables).

### What Pandas is used for
- Cleaning and preparing data
- Reading/writing data (CSV, Excel, databases, etc.)
- Filtering, sorting, and grouping data
- Performing calculations and statistics
- Handling missing values
- It’s powerful, easy to use, and integrates well with other tools like:  NumPy (for numerical operations), Matplotlib / Seaborn (for visualization), Machine learning libraries
### Core data structures
- Series → a single column (1D data)
- DataFrame → a table (rows and columns)
- different operations

### Pickling
- In pandas, pickle is a way to save Python objects (like DataFrames, lists, dictionaries, models, etc.) into a binary file so you can load them back later.

- Saving machine learning datasets
- Saving trained AI/ML models
- Faster loading than CSV
- Preserving exact Python object structure


```python
dfe.to_pickle("df_execl")
```

```python
dfe=pd.read_pickle("df_execl")
```

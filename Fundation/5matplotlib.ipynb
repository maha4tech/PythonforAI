```python
# compute the x and y coordinates for  point on a sine curve
x=np.arange(0,4 * np.pi, 0.1)
y=np.sin(x)
plt.title('sine wave ')

plt.plot(x,y)
plt.show()
```

```python
# subplot()
# compute the x an y coordinates for points on sinea d cosine curves

x=np.arange(0,5*np.pi, 0.1)
y_sin =np.sin(x)
y_cos =np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('consine')

plt.show()
```

```python
## bar plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y, align ='center')
plt.bar(x2,y2, color='g')
plt.ylabel('Y AXIS')
plt.xlabel('X AXIS')
plt.show()
```

```python
# histograms

a=np.array([22,87,20,43,56,73,25,20,79,32,27])
plt.hist(a)
plt.title("histogram")
plt.show()
```

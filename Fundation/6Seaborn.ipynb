# Distribution plots
- displot
- joinplot
- pairplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
df.dtypes
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

### Univariate analysis


```python
sns.jointplot(x='tip',y="total_bill",data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y="total_bill",data=df,kind='reg')
```

## pair plot
-  A pair plot is also know as a scetterplot , in which one variable in the same data row is matched with another variable's value, like this . pairs are just elaborations on this , showing all variables paired with all other variables

```python
sns. pairplot(df)
```

```python
sns. pairplot(df,hue='sex')
```

### Dist Plot
- dist plot us ot check the disttribution of the colums feature


```python
sns.displot(df['tip'], kde=True)
```

```python
sns.displot(df['tip'],kde=False, bins=10)
```

# Categorical Plots


 Seaborn also helps us in doing analysis on categorical Data points. in this section we will discuss about
- boxplot
- vioinplot 
- countplot
- barplot

```python
df.head()
```

```python
df.dtypes
```

```python
## count  plot

sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.countplot(x='day',data=df)
```

```python
# bar plot
sns.barplot(y='total_bill',x='sex',data=df)
```

### Box Plot
A box and whisker plot is a graph that presents information from a five -numbers summary


```python
sns.boxplot(x='smoker', y='total_bill',data=df, palette="rainbow")
```

```python
sns.boxplot(data=df, palette="rainbow")
```

```python

# catogorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker", data=df, palette="rainbow")
```

## Violin Plot
Violin plot help us to see both the distribution of data in terms of kernel desity estimation and the box plot


```python
sns.violinplot(x="total_bill",y="day",  hue="smoker", data=df, palette="rainbow")
```
